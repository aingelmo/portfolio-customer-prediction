```python
from pathlib import Path
import pandas as pd

DATA_DIR = Path("..") / "data"

marketing_data  = pd.read_csv(DATA_DIR / "marketing_campaign.csv", sep="\t")
```

```python
marketing_data
```

```python
marketing_data.columns
```

```python
from pycaret.regression import RegressionExperiment

model_data = marketing_data.filter(
    items=["Year_Birth", "Education", "Income", "MntWines"]
)

exp = RegressionExperiment()
exp.setup(model_data, target="MntWines", session_id=123, use_gpu=True)
```

```python
best = exp.compare_models()
```

```python
exp.plot_model(best, plot="feature")
```

```python
holdout_pred = exp.predict_model(best)
```

```python
holdout_pred
```

```python
to_predict = model_data.sample(20)
```

```python
to_predict.drop(columns=["MntWines"]).columns.to_list()
```

```python
exp.predict_model(best, data=to_predict)
```

```python
exp.save_model(best, 'my_first_pipeline')
```
